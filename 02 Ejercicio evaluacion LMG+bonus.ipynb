```python
#crear clase #TiendaOnline. Ejercicio Lourdes Moya

class TiendaOnline:

    def __init__(self):
        self.inventario = [] # es una lista de diccionarios[{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40}
        self.ventas_totales = 0.0 #guarda un float.
        self.clientes = {} #formato diccionario: {'Cliente1': {'email': 'cliente1@email.com', 'compras': []},

    def agregar_producto (self,nombre,precio,cantidad):
        '''Agrega un producto al inventario o actualiza su cantidad si ya existe'''

        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios.       
            if  elemento['nombre'] == nombre: #comparo si el nombre está ya creado.
                elemento['cantidad'] = cantidad #de ese producto, actualizo la cantidad por asignación directa.
                return print("Se ha actualizado la cantidad") #uso este return en el if para que salga de la función y termina.
                
        self.inventario.append({"nombre":nombre, "precio":precio,"cantidad":cantidad}) #NO ENTRA EN EL IF porque no existe, y la creo

    def ver_inventario (self):
        '''Muestra el inventario de productos con sus detalles.'''

        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios.       
            print(f"Nombre: {elemento["nombre"]}, Precio: ${elemento["precio"]}, Cantidad: {elemento["cantidad"]}")

    def buscar_producto(self,nombre):
        '''Busca un producto en el inventario por nombre y muestra sus detalles si se encuentra'''

        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios.       
            if  elemento['nombre'] == nombre: #comparo si el producto está .
                return print(f"Nombre: {elemento["nombre"]}, Precio: ${elemento["precio"]}, Cantidad: {elemento["cantidad"]}")
            
        print("Producto no encontrado.")
                
    def actualizar_stock(self, nombre, cantidad):
        '''Actualiza el stock de un producto en el inventario'''

        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios.       
            if  elemento['nombre'] == nombre: #comparo si el nombre está ya creado.
                elemento['cantidad'] = cantidad #de ese producto, actualizo la cantidad por asignación directa.
                return print("La cantidad ha sido actualizada en el stock")
        
        print("El producto no está en el inventario.")

    def eliminar_producto(self, nombre):
        '''Elimina un producto del inventario por nombre.'''

        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios.       
            if  elemento['nombre'] == nombre: #comparo si el nombre está ya creado.
                self.inventario.remove(elemento)
                return print (f"El elemento {nombre} ha sido eliminado")
            
        print("El producto no está en el inventario")

    def calcular_valor_inventario (self):
        '''Calcula y muestra el valor total del inventario.'''

        suma_total = 0 #defino la variable donde voy a guardar el cálculo.
        for elemento in self.inventario: #por cada elemento (diccionario) en la lista de diccionarios. 
            suma_total += (elemento["cantidad"]*elemento["precio"]) #de cada elemento, que multiplique cantidad y precio.
        

        return print(f"El total del inventario asciende a ${suma_total}")
    
    def agregar_cliente (self,nombre,email):
        '''Agrega un nuevo cliente al registro de clientes.'''

        for value in self.clientes.keys(): #como solo necesito comparar el valor de la key para comprobar si ese cliente ya existe.
            
            if value == nombre: #compruebo si existe el cliente sacandome las keys del diccionario.
                return print("Existe ese cliente ya.")
            
        self.clientes.update({nombre:{"email":email}}) #uso setdefault para crear


    def ver_clientes(self):
        '''Muestra la lista de clientes registrados con sus nombres y correos electrónicos.'''

        print("A continuación se muestra cada cliente seguido de su e-mail:""\n")

        print("Nombre    : E-mail")
        print("-----------------")
        for k,v in self.clientes.items(): 
            print(k,":",v["email"]) #aqui voy imprimiendo la clave principal (cliente) con su value (email), aun no está compra.

    def realizar_compra(self):
        '''Permite a un cliente realizar una compra seleccionando productos del inventario. '''

        costo_compra =0 #variable para calcular el total del carrito
```
